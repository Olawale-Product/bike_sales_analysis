### Numerical Analysis and  Visualization
### we analyze the unit cost column


```python
sales.Unit_Cost.describe()
```

```python
sales.Unit_Cost.mean()
```

```python
sales.Unit_Cost.median()
```

```python
sales.Unit_Cost.plot(kind ="box", vert=False,figsize =(14,8))
```

```python
sales.Unit_Cost.plot(kind ="density",figsize =(14,8))
```

```python
ax =sales.Unit_Cost.plot(kind ="density",figsize =(14,8))
ax.axvline(sales.Unit_Cost.mean(),linewidth=2, color="r")
ax.axvline(sales.Unit_Cost.median(), linewidth=2, color="b")
```

```python
ax = sales["Unit_Cost"].plot(kind = "hist", figsize=(14,8))
ax.set_xlabel(" Amount in dollars($) ",fontsize =16)
ax.set_ylabel("Number of Sales", fontsize =12)
```

## Categorical Analysis and Visualisation
#### we analyze the age group column

```python
sales["Age_Group"].value_counts()
```

```python
ax = sales["Age_Group"].value_counts().plot(kind="pie", figsize=(10,10));
ax.set_xlabel("Number of Sales",color="white", fontsize=16 ); 
```

```python
ax = sales["Age_Group"].value_counts().plot(kind="bar", figsize=(14,10));
ax.set_ylabel("Number of Sales", fontsize=16 ,labelpad=40.0); 
```
